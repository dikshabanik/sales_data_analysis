```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('sales_data.csv')
```

Data Cleaning

```python
data['Date'] = pd.to_datetime(data['Date'])  #converting date column to datetime format
```

Calculating the Total Sales and the Average Sales per transaction

```python
data['Total_Sales'] = data['Price'] * data['Quantity']
total_sales = data['Total_Sales'].sum()
average_sales = data['Total_Sales'].mean()

print(f"Total Sales: {total_sales}")
print(f"Average Sales per Transaction: {average_sales:.2f}")
```

Identifying top-selling products

```python
#basically we are clubbing the same products together first and then sorting them in descending order
#head(n) is used to return the first n datasets
top_products = data.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False).head(5)
print("Top 5 Best-Selling Products:")
print(top_products)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
top_products.plot(kind='bar', color='skyblue')
plt.title('Top 5 Best-Selling Products')
plt.ylabel('Total Sales')
plt.xlabel('Product')
```

```python
data.loc[data['Transaction_ID'] == 1, 'Price'] = 15000
data.loc[data['Transaction_ID'] == 5, 'Price'] = 15000
data.loc[data['Transaction_ID'] == 4, 'Price'] = 5000
data.loc[data['Transaction_ID'] == 3, 'Price'] = 750

data.to_csv('sales_data.csv', index=False)  #to save the changes made in the csv file

print("Prices updated successfully!")
```

```python
data2 = pd.read_csv('sales_data.csv')  #we have to read the file again
```

```python
plt.subplot(1, 2, 1)
data2.groupby('Product')['Price'].sum().sort_values(ascending=False).plot(kind='bar', color='pink')
plt.title('Total Sales by Product (after updating the prices)')
plt.xlabel('Product')
plt.ylabel('Total Sales')
```
